# Analyse du tourisme Vancouver

## 1 - Vancouver 

La première question qu'on peut se poser c'est si le tourisme augmente et églament si l'on peut imputer cela au tourisme 

```python
import pandas as pd
import matplotlib.pyplot as plt 

vancouver_visits = pd.read_csv('datasets_full/Vancouver/vancouver_visits.csv', delimiter = ',')
```

```python
vancouver_visits
```

```python
vancouver_visits.plot(x='year', y='value')
```

```python
vancouver_visits.groupby(['month','year']).sum()
```

```python
vancouver_visits[vancouver_visits['year'] == 1994].plot(y='value')
```

```python
vancouver_visits[vancouver_visits['year'] == 2000].plot(y='value')
```

```python
vancouver_visits[vancouver_visits['year'] == 2010].plot(y='value')
```

#### Globalement on observe un pic de tourisme au mois de Mai, Juin, Juillet, Août (et une tendance similaire chaque année)

#### Regardons les mois de Février (JO se déroulant du 12 au 28 février)

```python
vancouver_visits[vancouver_visits['month']==2]

vancouver_visits[vancouver_visits['month']==2].plot(y='value', x='year')
```

#### On voit donc un pic en 2010 pour le moi de février.

#### Que se passe-t-il pour les autres mois ? 

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
vancouver_visits.groupby(['year','month']).sum()['value'].unstack().plot(ax=ax)

plt.axvline(x=2010, ymin=0, ymax=1500000)

plt.show()
```

```python
# plot data
fig, ax = plt.subplots(figsize=(15,7))
# use unstack()
vancouver_visits.groupby(['month','year']).sum()['value'].unstack().plot(ax=ax)
```

```python
vancouver_visits.groupby(['year']).sum()['value'].plot()
```

#### On observe ainsi une petit eaugmentation du tourisme le mois de Février en 2010 par rapport aux autres années 

#### Une (trop) légère augmentation du tourisme en 2010 dans l'année



```python
vancouver_visits.groupby(['year']).sum()['value']
```

```python
vancouver_visits.groupby(['year']).sum()['value'].to_csv('vancouver_visits_year.csv')
```
